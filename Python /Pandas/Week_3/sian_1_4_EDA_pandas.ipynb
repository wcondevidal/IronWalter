### State by State analysis

```python
data['state'].value_counts()
```

```python
#the State field needs cleaning a little- as we did yesterday with gender

def clean_state(x):
    if x in ['Cali', 'California']:
        return 'CA'
    elif x == 'Arizona':
        return 'AZ'
    elif x == 'Tennessee':
        return 'TN'
    else: 
        return x
```

```python
data['state']=list(map(clean_state,data['state']))
```

```python
data['state'].value_counts()
```

```python
#create a grouped dataframe, like a pivot - summarising the data by state 
groupby = data.groupby('state', axis=0)
```

```python
#take a look at the means of all data points by state 
groupby.mean()
```

```python
#create a data frame which looks at avg gifts by state 
summ=data.groupby('state')['avggift'].mean().sort_values()
```

```python
#we can plot that
ax=summ.plot.barh(x='avggift', y=['state'], color='green')
```

### Dive into correlation

```python
#review correlation of two selected variables 
plt.scatter(x=data['avggift'], y=data['ic3'])
plt.show()
```

```python
#bring in the scipy library - The scientific python library
# an open-source Python library which is used to solve scientific and mathematical problems. 
# It is built on the NumPy extension and allows the user to manipulate and visualize data.
import scipy.stats as stats
```

```python
x=data['avggift']
y=data['ic3']
stats.pearsonr(x, y) # gives correlation coefficient and the p value 
```

```python
stats.spearmanr(x, y)
```

```python
stats.kendalltau(x, y)
```

```python
#create a correlation matrix of numerical columns 

corr_data = data.drop(['id','state', 'gender','domain'], axis=1)
correlations_matrix = corr_data.corr()
correlations_matrix
```

```python
#create a heatmap 
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
#use the mask technique to make it easier to read 
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```
