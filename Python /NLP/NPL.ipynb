```python
from nltk.tokenize import word_tokenize
from nltk.text import Text
```

```python
# def variable 'textfile' doc2txt.process() 
# use nltk to separate the individual words - word_tokenise 
textfile=docx2txt.process("Documento sin título.docx")
```

```python
textfile
```

```python
tokens = word_tokenize(textfile)
```

```python
tokens[:11]
```

```python
len(tokens)
```

```python
fdist1=nltk.FreqDist(tokens)
```

```python
word_freq= dict((word,freq) for word, freq in fdist1.items() if not word.isdigit())
word_freq
```

```python
sorted=sorted(word_freq.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted
```

```python
# remove punctuation, spaces, numbers 

clean_words=[word for word in tokens if word.isalpha()]

clean_words[:10]
```

```python
lower_words=[word.lower() for word in clean_words]

lower_words[:10]
```

```python
#to remove stopwords - create your own stop word list ? the words which have no importance to your project 
# or you can use a stopword dictionary 

from nltk.corpus import stopwords
```

```python
stop_words=stopwords.words('spanish')
```

```python
print(stop_words)
```

```python
words_without_stop=[word for word in lower_words if not word in stop_words]
```

```python
len(words_without_stop)
```

```python
fdist2=nltk.FreqDist(words_without_stop)
```

```python
fdist2.most_common(20)
```

```python
# convert words into nltk language/ data type 

textlist=Text(words_without_stop)
```

```python
type(textlist)
```

```python
#words that frequently appear together (phrases) 
textlist.collocations()
```

```python
# find fragments containing key words - annother index is creared - so searching is faster next time 

textlist.concordance("derecho")
```

```python
#lexical dispersion plot - analyses appearances of words 

textlist.dispersion_plot(["ley", "reparar", "aparatos", "propuesta", "fabricantes"])
```
