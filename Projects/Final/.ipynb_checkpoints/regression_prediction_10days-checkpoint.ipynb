```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
df = pd.read_csv ("cases_keywords_complete_V2_10days - Hoja 1.csv")
```

```python
df.head()
```

```python
data = df.drop(["fecha","Index","suma_keys_sintomas","suma_keys_preguntas","casos_5"],axis=1)
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.info
```

```python
y = data["casos_10"]
```

```python
X = data.drop(["casos_10"],axis=1)
```

```python
lm = linear_model.LinearRegression()
```

```python
line = lm.fit(X,y)
lm.score(X,y)
```

```python
predictions1 = lm.predict(X)
predictions1
```

```python
data["prediction"] = predictions1
```

```python
data["fecha"] = df["fecha"]
```

```python
predictions1 = predictions1.astype(int)
```

```python
data.tail(40)
```

```python
r2_score(y, predictions1)
```

```python
data1 = data.drop(columns = ["prediction", "fecha"])
data1.head()
```

```python
train = data1.iloc[:170]
train
```

```python
test = data1.iloc[170:]
test
```

```python
y_train = train["casos_10"]
X_train = train.drop(["casos_10"],axis=1)

y_test = test["casos_10"]
X_test = test.drop(["casos_10"],axis=1)
```

```python
lm=linear_model.LinearRegression()
model=lm.fit(X_train,y_train) # teach the model to find the line of best fit 
predictions=lm.predict(X_test) # run predictions based on X_test (we cannot see y_test)
r2_score(y_test, predictions)#c
```

```python
model.fit(X_train, y_train)
```

```python
train_pred = model.predict(X_train).astype(int)
train_pred
```

```python
test_pred = model.predict(X_test).astype(int)
test_pred
```

```python
total_pred = np.concatenate([train_pred, test_pred])
```

```python
r2_score(y, total_pred)
```

```python
mean_squared_error(y, total_pred, squared=False)
```
